```python
import pandas as pd
```

```python
# read from csv

articoli = pd.read_csv('Lista Articoli.csv', decimal=',')
storico = pd.read_csv('Storico Filtrato.csv', decimal=',')
```

# Processing

```python
storico.head()
```

```python
df = storico.rename(columns={'Unnamed: 1':'anno', 'Unnamed: 2':'mese', 'Unnamed: 3':'macchina'})
```

```python
# TODO: make anno, mese utilizzabili

df = df.drop(['N° MC','anno', 'mese'], axis=1)
df = df.drop([0, 1])
```

```python
df.dropna(subset=['macchina'], inplace=True)
df['macchina'] = df['macchina'].astype(int)  
```

```python
df.head()
```

```python
df['data ora partenza'] = pd.to_datetime(df['data ora partenza'], format='%d/%m/%Y %H:%M:%S', errors='coerce')
df['data ora fine'] = pd.to_datetime(df['data ora fine'], format='%d/%m/%Y %H:%M:%S', errors='coerce')
```
